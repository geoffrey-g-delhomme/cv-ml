# Activation

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4,4,100, endpoint=True)

def plot(x, y):
    plt.plot(x, y)
    plt.grid(True)
```

## tanh

$$f(x) = \tanh(x)$$

```python
def f(x):
    return np.tanh(x)

plot(x, f(x))
```


## sigmoid

$$f(x) = \left ( 1+e^{-x} \right ) ^{-1}$$

## 

```python
def f(x):
    return np.power((1 + np.exp(-x)),-1)

plot(x, f(x))
```

## Rectified Linear Unit (ReLU)

$$f(x) = \left ( 1+e^{-x} \right ) ^{-1}$$

Introduced in [ImageNet Classification with Deep Convolutional Neural Networks](https://papers.nips.cc/paper_files/paper/2012/file/c399862d3b9d6b76c8436e924a68c45b-Paper.pdf):
- Deep convolutional neural networks with ReLUs train several times faster than their equivalents with `tanh` units.

```python
def f(x):
    return np.clip(x, 0, a_max=None)

plot(x, f(x))
```
